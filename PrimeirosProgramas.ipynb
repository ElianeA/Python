```
genero = str(input("Digite F para feminino e M para masculino: ").upper())
if genero == 'F':
  print(f"Feminino")
elif genero == 'M':
  print(f"Masculino")
else:
  print(f"Genero invalido")
```

```
vogal = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
letra = str(input("Digite F para feminino e M para masculino: "))
if letra in vogal:
  print(f"A letra {letra} é uma vogal")
else:
  print(f"A letra {letra} é uma consoante")
```
